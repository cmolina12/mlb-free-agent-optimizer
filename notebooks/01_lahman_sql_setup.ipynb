```python
import sqlite3
import pandas as pd
from pathlib import Path

# Paths
RAW = Path("../data/raw/lahman/")
DB_PATH = Path("../data/processed/db/lahman.db")

# Create processed/db directory if missing
DB_PATH.parent.mkdir(parents=True, exist_ok=True)

# Connect to SQLite DB (it will be created if it doesn't exist)
conn = sqlite3.connect(DB_PATH)
conn
```

```python
tables_to_import = [
    "People",
    "Batting",
    "Pitching",
    "Teams",
    "Salaries"  # comment this out if you don't have it
]

for table in tables_to_import:
    file_path = RAW / f"{table}.csv"
    if file_path.exists():
        print(f"Loading {file_path}...")
        df = pd.read_csv(file_path)
        df.to_sql(table, conn, if_exists="replace", index=False)
        print(f"Imported {table}: {len(df)} rows")
    else:
        print(f"{table}.csv not found - skipping")
```
