```python
import pandas as pd
from pathlib import Path

from pybaseball import statcast_batter_exitvelo_barrels

# Paths
PROCESSED = Path("../data/processed/features/")
PROCESSED.mkdir(parents=True, exist_ok=True)

YEARS = [2020, 2021, 2022, 2023, 2024, 2025]
```

## Download Statcast hitter stats for each year

```python
all_years = []

for year in YEARS:
    print(f"Downloading Statcast hitting data for {year}...")
    df_year = statcast_batter_exitvelo_barrels(year)
    df_year["year"] = year
    all_years.append(df_year)

len(all_years)
```

## Combine and save to CSV

```python
statcast_hitters = pd.concat(all_years, ignore_index=True)

print(statcast_hitters.shape)
statcast_hitters.head()
```

```python
output_path = PROCESSED / "statcast_hitting_2020_2025.csv"
statcast_hitters.to_csv(output_path, index=False)

output_path
```
